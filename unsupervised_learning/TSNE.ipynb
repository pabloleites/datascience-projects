```python
from sklearn.manifold import TSNE
import pandas as pd
import matplotlib.pyplot as plt
```

```python
samples = pd.read_csv('datasets/seeds.csv',header=None)
varieties = samples[7]
samples = samples.drop(7,axis=1)
samples
```

# t-SNE
is a statistical method for visualizing high-dimensional data by giving each datapoint a location in a two or three-dimensional map.

```python
model = TSNE(learning_rate=200)
tsne_features = model.fit_transform(samples)

xs = tsne_features[:,0]
ys = tsne_features[:,1]
# Scatter plot, coloring by variety_numbers
plt.scatter(xs, ys, c=varieties)
plt.show()
```
