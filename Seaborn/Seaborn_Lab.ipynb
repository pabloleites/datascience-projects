```python
%matplotlib inline
# Import Matplotlib, Pandas, and Seaborn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

plt.rcParams['figure.figsize'] = [8, 6]
csv_filepath = "young-people-survey-responses.csv"

# Create a DataFrame from csv file
df = pd.read_csv(csv_filepath)

# Create a count plot with "Spiders" on the x-axis
sns.countplot(x="Spiders", data=df)

# Display the plot
plt.show()
```

```python
student_data = pd.read_csv("student-alcohol-consumption.csv")
sns.scatterplot(x="absences", y="G3", 
                data=student_data, 
                hue="location", hue_order=['Rural','Urban'])

# Show plot
plt.show()
```

```python
# Create a dictionary mapping subgroup values to colors
palette_colors = {"Rural": "green", "Urban": "blue"}

# Create a count plot of school with location subgroups
sns.countplot('school', data=student_data, hue='location', palette=palette_colors)

# Display plot
plt.show()
```

```python
# Adjust further to add subplots based on family support
sns.relplot(x="G1", y="G3", 
            data=student_data,
            kind="scatter", 
            col="schoolsup",
            col_order=["yes", "no"], row='famsup', row_order=["yes","no"])

# Show plot
plt.show()
```

```python
mpg = pd.read_csv("mpg.csv")
sns.relplot(x="horsepower", y="mpg", 
            data=mpg, kind="scatter", 
            size="cylinders", hue="cylinders")

# Show plot
plt.show()
```

```python
# Create a scatter plot of acceleration vs. mpg

sns.relplot(kind="scatter", data=mpg, x="acceleration", y="mpg", hue="origin", style="origin")


# Show plot
plt.show()
```

```python
# Create line plot

sns.relplot(data=mpg, x="model_year", y="mpg", kind="line")

# Show plot
plt.show()
```

```python
# Make the shaded area show the standard deviation
sns.relplot(x="model_year", y="mpg",
            data=mpg, kind="line", ci="sd")
# Show plot
plt.show()
```

```python
# Add markers and make each line have the same style
sns.relplot(x="model_year", y="horsepower", 
            data=mpg, kind="line", 
            ci=None, style="origin", 
            hue="origin", markers=True, dashes=False)

# Show plot
plt.show()
```

```python
# import data
survey_data=pd.read_csv("young-people-survey-responses.csv")

# create categorical values
survey_data["Age Category"] = pd.cut(survey_data['Age'], [0,21,np.inf], labels=['less than 21','21+'])
survey_data["Interested in Math"] = survey_data['Mathematics'] > 3


# Create column subplots based on age category
sns.catplot(col="Age Category", y="Internet usage", data=survey_data,
            kind="count")

# Show plot
plt.show()
```

```python
# Create a bar plot of interest in math, separated by gender
sns.catplot(kind='bar', data=survey_data, x="Gender", y="Interested in Math")

# Show plot
plt.show()
```

```python
# Turn off the confidence intervals
sns.catplot(x="study_time", y="G3",
            data=student_data,
            kind="bar",
            order=["<2 hours", 
                   "2 to 5 hours", 
                   "5 to 10 hours", 
                   ">10 hours"], ci=None)

# Show plot
plt.show()
```

```python
# Specify the category ordering
study_time_order = ["<2 hours", "2 to 5 hours", 
                    "5 to 10 hours", ">10 hours"]

# Create a box plot and set the order of the categories
sns.catplot(kind='box', data=student_data, x="study_time", y="G3", order=study_time_order)
plt.xticks(rotation=90)
# Show plot
plt.show()
```

```python
# Create a box plot with subgroups and omit the outliers

sns.catplot(kind='box', data=student_data, x="internet", y="G3", hue="location", sym="")

# Show plot
plt.show()
```

```python
# Set the whiskers at the min and max values
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box",
            whis=[0, 100])

# Show plot
plt.show()
```

```python
# Remove the lines joining the points
sns.catplot(x="famrel", y="absences",data=student_data, kind="point", capsize=0.2, join=False)
            
# Show plot
plt.show()
```

```python

# Plot the median number of absences instead of the mean
sns.catplot(x="romantic", y="absences",
            data=student_data,
            kind="point",
            hue="school",
            ci=None, estimator=np.median)

# Show plot
plt.show()
```

```python
# Create scatter plot
g = sns.relplot(x="weight", 
                y="horsepower", 
                data=mpg,
                kind="scatter")

# Add a title "Car Weight vs. Horsepower"
g.fig.suptitle("Car Weight vs. Horsepower")
#Rotate x labels 90 degrees
plt.xticks(rotation=90)
# Show plot
plt.show()
```

```python
# Create line plot
g = sns.lineplot(x="model_year", y="mpg", 
                 data=mpg,
                 hue="origin", ci=None)

# Add a title "Average MPG Over Time"
g.set_title("Average MPG Over Time")

# Add x-axis and y-axis labels

g.set(xlabel="Car Model Year", ylabel="Average MPG")

# Show plot
plt.show()
```
